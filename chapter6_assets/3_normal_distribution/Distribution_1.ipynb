```python
df_mean = round(df.height.mean(),2)
df_std = round(df.height.std(),2)

df_mean,df_std

```

```python
import seaborn as sns
from matplotlib import pyplot as plt

sns.histplot(df.height, kde = True)

# Add vertical lines for mean, mean + std, and mean - std
plt.axvline(df_mean, color='red', linestyle='--', label='Mean')

plt.axvline(df_mean + df_std, color='green', linestyle='--', label='Mean + 1 Std')
plt.axvline(df_mean - df_std, color='green', linestyle='--', label='Mean - 1 Std')

plt.axvline(df_mean + df_std*2, color='orange', linestyle='--', label='Mean + 1 Std')
plt.axvline(df_mean - df_std*2, color='orange', linestyle='--', label='Mean - 1 Std')

plt.axvline(df_mean + df_std*3, color='purple', linestyle='--', label='Mean + 1 Std')
plt.axvline(df_mean - df_std*3, color='purple', linestyle='--', label='Mean - 1 Std')
```

```python
import pandas as pd
df = pd.read_csv('heights.csv')

df_mean = round(df.height.mean(),2)
df_std = round(df.height.std(),2)

left1 = round(df_mean - df_std, 2)
right1 = round(df_mean + df_std, 2)

left1,right1
```

```python
left2 = round(df_mean-2*df_std, 2)
right2 = round(df_mean+2*df_std, 2)

left2,right2
```

```python
left3 = round(df_mean-3*df_std, 2)
right3 = round(df_mean+3*df_std, 2)

left3,right3
```

```python
df[(df.height>left1) & (df.height<right1)].shape
# 64% data is coming under the green lines
```

```python
df[(df.height>left2) & (df.height<right2)].shape
# 97% data is coming under the yellow lines
```

```python
df[(df.height>left3) & (df.height<right3)].shape
# 99% data is coming under the purple lines
```
