```python
import pandas as pd
df = pd.read_csv("heights_with_outliers.csv")
df.head(5)
```

```python
df_mean = round(df.height.mean(),2)
df_std = round(df.height.std(),2)

df_mean,df_std
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

sns.histplot(df.height, kde = True)

# Add vertical lines for mean, mean + std, and mean - std
plt.axvline(df_mean, color='red', linestyle='--', label='Mean')

plt.axvline(df_mean + df_std, color='green', linestyle='--', label='Mean + 1 Std')
plt.axvline(df_mean - df_std, color='green', linestyle='--', label='Mean - 1 Std')

plt.axvline(df_mean + df_std*2, color='orange', linestyle='--', label='Mean + 1 Std')
plt.axvline(df_mean - df_std*2, color='orange', linestyle='--', label='Mean - 1 Std')

plt.axvline(df_mean + df_std*3, color='purple', linestyle='--', label='Mean + 1 Std')
plt.axvline(df_mean - df_std*3, color='purple', linestyle='--', label='Mean - 1 Std')
```

```python
lower = df.height.mean() - df_std*3
upper = df.height.mean() + df_std*3
lower, upper
```

```python
outlier = df[(df.height < lower) | (df.height > upper)]
outlier
```

```python
df_no_outliers = df[(df.height>lower)&(df.height<upper)]
df_no_outliers.describe()
```

```python
sns.histplot(df_no_outliers, kde=True)
```
